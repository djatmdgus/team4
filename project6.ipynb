```python
def lav(x):
    if x == '불안':
        return int(0)
    if x == '분노':
        return int(1)
    if x == '슬픔':
        return int(2)
    if x == '기쁨':
        return int(3)
    if x == '당황':
        return int(4)
    if x == '상처':
        return int(5)
    if x == '불안 ':
        return int(0)
    if x == '기쁨 ':
        return int(3)

df['감정_대분류'] = df['감정_대분류'].apply(lav)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token = Tokenizer()
token.fit_on_texts(df['사람문장1'])             # 토큰화
word_index = token.word_index           # 인덱스 별 단어

print('총단어 갯수',len(word_index))
word_index
```

```python
df['사람문장1'] = token.texts_to_sequences(df['사람문장1'])     # 시퀸스 형태로 변환 ( 각 단어의 인덱스 )
df
```

```python
sequences_len = [len(x) for x in df['사람문장1'] ]

print('문장 최대 길이',max(sequences_len))
print('문장 최소 길이',min(sequences_len))
print('문장 평균길이',sum(sequences_len) / len(sequences_len))

```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['사람문장1'], df['감정_대분류'], test_size=0.25, stratify=df['감정_대분류'], random_state=2)
```

```python
y_train.isnull().sum()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_len = 28    # 문장 최대길이

pad_x_train = pad_sequences(x_train, maxlen=max_len, padding='pre')   # 패딩처리 ( 문장 길이 맞춤 )
pad_x_test = pad_sequences(x_test, maxlen=max_len, padding='pre')
```

```python
pad_x_train.shape, pad_x_test.shape
```

```python
pad_x_train[0]
```

```python
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, Embedding, Dense, Bidirectional, Dropout
```

```python
size = len(word_index) + 1  #인풋총 단어갯수 0부터니 1더해서,, 전체 단어 갯수랑 비슷하면 학습시 과적합 발생 이유 모름 
dim = 150 #줄일 차원? 단어?
max_length = 28   #입력 단어 
```

```python
model = Sequential()
model.add(Embedding(2000, dim, input_length=max_length))
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(128, return_sequences=True)))
model.add(Bidirectional(LSTM(128)))
model.add(Dropout(0.25))
model.add(Dense(6, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='sparse_categorical_crossentropy', #sparse_categorical_crossentropy
             optimizer='adam',
             metrics=['accuracy'])
```

```python
model.fit(pad_x_train, y_train, 
            epochs=8,
            batch_size=128,
            validation_data=(pad_x_test, y_test))
```

```python
import pickle
```

```python
with open('model.pkl','wb') as pickle_file:
    pickle.dump(model, pickle_file)
```
